Apple or banana Classification(With KNN)

- Apple width 6
- Apple length6
- Banana width 4
- Banana length 8

Apple = 0

Banana = 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 100
apple_width = np.random.normal(6, 1, n)
apple_length = np.random.normal(6, 1, n)
banana_width = np.random.normal(4, 1, n)
banana_length = np.random.normal(8, 1, n)
```

```python
plt.scatter(apple_width, apple_length, c='r', marker='.', label='apple')
plt.scatter(banana_width, banana_length, c='orange', marker='.', label='banana')
plt.title('Apple and Banana')
plt.xlabel('width')
plt.ylabel('length')
plt.legend()
plt.show()
```

```python
width = np.concatenate((apple_width, banana_width))
length = np.concatenate((apple_length, banana_length))

X_train = np.array([width,length]).T # T = transpose rows and cols
X_train
```

```python
Y_apples = np.zeros(100, dtype=int)
Y_bananas = np.ones(100, dtype=int)
Y_train = np.concatenate((Y_apples, Y_bananas))
Y_train
```

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2)** 2))
```

```python
new_fruit = np.array([5.5, 7])

distances = []
for fruit in X_train:
    d = euclidean_distance(new_fruit, fruit)
    distances.append(d)
distances
```

```python
k = 5
nearest_neighbors =  np.argsort(distances)[0:k]
nearest_neighbors
```

```python
Y_train[nearest_neighbors]
```

```python
result = np.bincount(Y_train[nearest_neighbors])
result
```

```python
if np.argmax(result) == 0:
    print("It's an apple")
else:
    print("It's a banana")
```
