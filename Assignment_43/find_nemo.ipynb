```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

from knn import KNN
```

```python
nemo = cv.imread('Nemo.jpg')
nemo = cv.resize(nemo, (0, 0), fx=0.25, fy=0.25)
nemo = cv.cvtColor(nemo, cv.COLOR_BGR2RGB)
```

```python
pixels_list_rgb = nemo.reshape(-1, 3)
pixels_list_rgb.shape
```

```python
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')
ax.scatter(pixels_list_rgb[:, 0], pixels_list_rgb[:, 1], pixels_list_rgb[:, 2], c=pixels_list_rgb / 255, marker="o")
ax.set_xlabel('Red')
ax.set_ylabel('Green')
ax.set_zlabel('Blue')
plt.show()
```

```python
nemo_hsv = cv.cvtColor(nemo, cv.COLOR_BGR2HSV)
pixels_list_hsv = nemo_hsv.reshape(-1, 3)
ax = plt.figure(figsize=(8, 8)).add_subplot(projection='3d')

ax.scatter(pixels_list_hsv[:, 0], pixels_list_hsv[:, 1], pixels_list_hsv[:, 2], c=pixels_list_rgb / 255, marker="o")
ax.set_xlabel('Hue')
ax.set_ylabel('Saturation')
ax.set_zlabel('Value')
plt.show()
```

```python
light_orange = (100, 50, 130)
dark_orange = (115, 255, 255)

orange_mask = cv.inRange(nemo_hsv, light_orange, dark_orange)
plt.imshow(orange_mask, cmap='gray')
plt.show()
```

```python
orange_mask.shape
```

```python
light_withe = (75, 0, 150)
dark_withe = (130, 50, 253)

white_mask = cv.inRange(nemo_hsv, light_withe, dark_withe)
plt.imshow(white_mask, cmap='gray')
plt.show()
```

```python
final_mask = white_mask + orange_mask
plt.imshow(final_mask, cmap='gray')
plt.show()
```

```python
final_result = cv.bitwise_and(nemo, nemo, mask=final_mask)
plt.imshow(final_result, cmap='gray')
plt.show()
```

Now we use these informations for our KNN

```python
X_train = pixels_list_hsv / 255
X_train.shape
```

```python
Y_train = final_mask.reshape(-1,) // 255
Y_train.shape
```

```python
Y_train[20000:20100]
```

```python
knn = KNN(k=3)
knn.fit(X_train, Y_train)
```

```python
new_pixel = np.array([10, 0, 0]) // 255
knn.predict([new_pixel])
```
