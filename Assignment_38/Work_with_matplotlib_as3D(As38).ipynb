```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ballons properties
b_x = np.random.normal(26, 5, 100)
b_y = np.random.normal(25, 5, 100)
b_z = np.random.normal(31, 5, 100)

# Melons properties
m_x = np.random.normal(25, 5, 100)
m_y = np.random.normal(27, 5, 100)
m_z = np.random.normal(24, 5, 100)
```

```python
# For drawing a 3D plot
ax = plt.figure(figsize=(8, 5)).add_subplot(projection='3d')


# Setting ranges for vector numbers
ax.set_xlim(10, 42)
ax.set_ylim(10, 42)
ax.set_zlim(10, 42)

# Setting labels
ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
ax.set_zlabel("Z Label")

# for define witch you want to draw and how you want draw it, you can use scatter or plot
ax.scatter(b_x, b_y, b_z, c="r", label='Ballons shapes(x, y, z)')
# ax.scatter(m_x, m_y, m_z, 'g^', label='Melons shapes(x, y, z)')
# If you want change the visual shape of data(3rd argument), you must use plot instead of scatter
ax.plot(m_x, m_y, m_z, 'c^', label='Melons shapes(x, y, z)')

# For showing labels as legend
ax.legend()

# This method used when you want change your sight manually
ax.view_init(elev=30, azim=-60, roll=0)

# For showing all datas at final
# plt.figure(figsize=(20, 20))
# plt.show()
```
